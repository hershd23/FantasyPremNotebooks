## Goals plot

```python
def color_array(pos_array):
    color_arr = []
    for i in range(len(pos_array)):
        if pos_array[i] in ['F', 'F S']:
            color_arr.append('gold')
        elif pos_array[i] in ['F M S', 'F M'] :
            color_arr.append('dodgerblue')
        else:
            color_arr.append('xkcd:salmon')
            
    return np.array(color_arr)
```

```python
def goals_perc(x, y):
    if x > 10:
        return True
    return False
```

```python
plt.style.use("dark_background") ##I used a custom stylesheet in the original viz for the background color and fonts (which is Source Sans Pro)
fig, ax = plt.subplots(figsize=(20,20))
fg_color = 'white'
plt.title('PL 19-20, Min 5 Goals', color=fg_color, size = 36)
n=len(goals_df['goals']) ##number of points
xs, ys = goals_df['goals'].values , goals_df['xG'].values
colors = color_array(goals_df['position'].values)
ax.scatter(xs, ys, s=110, c=colors, alpha=.3, ec="k", linewidth=.4, hatch=r"\\\\")

selected = []
k = 0
for x,y,lbl in zip(goals_df['goals'].values , goals_df['xG'].values, goals_df['player_name']):
    label = lbl
    # this method is called for each point
    if goals_perc(x, y):
        selected.append(k)
        plt.annotate(label, # this is the text
                     (x,y), # this is the point to label
                     textcoords="offset points", # how to position the text
                     xytext=(6,6), # distance from text to points (x,y)
                     ha='left',
                     fontsize=14
                    ) # horizontal alignment can be left, right or center
    k+=1

##highlight certain selected points
sel_ind = np.array(selected) ##select certain indices
sel_x = xs[sel_ind]
sel_y = ys[sel_ind]
sel_cols = colors[sel_ind]
ax.scatter(sel_x, sel_y, s=120,c=sel_cols , alpha=.95, ec="white", linewidth=.6,) ##changed the alpha and removed the hatch for a more focused appearance

plt. plot([0,25], [0,25], linewidth=2.0, color='red')

##axes cosmetics
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(axis="both", which = "both", size=5, bottom=False, top=False, labelbottom=True, right=False, labelright=False, left=False, labelleft=True, labeltop=False)
ax.set_xlabel('Goals', color=fg_color, size = 16)
ax.set_ylabel('xG', color=fg_color, size = 16)
ax.text(0.95, 0.01, 'STRIKERS',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='gold', fontsize=28)
ax.text(0.95, 0.05, 'MIDFIELDERS',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='xkcd:salmon', fontsize=28)
ax.text(0.95, 0.09, 'WINGERS',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='dodgerblue', fontsize=28)
plt.savefig('./figs/pl_g_xg.png')
plt.show()
```

```python
minutes_df = player_df[player_df['time'] > 1500]
minutes_df
```

```python
minutes_df['npxG_90'] = (minutes_df['npxG']/minutes_df['time'])*90
minutes_df['xGB_90'] = (minutes_df['xGBuildup']/minutes_df['time'])*90
minutes_df['xGC_90'] = (minutes_df['xGChain']/minutes_df['time'])*90
```

```python
minutes_df = minutes_df[minutes_df['position'].isin(['M', 'M S'])]
minutes_df = minutes_df[~minutes_df['player_name'].isin(['Roberto Firmino', 'Michail Antonio'])]
minutes_df
```

```python
def color_array(pos_array):
    color_arr = []
    for i in range(len(pos_array)):
        if pos_array[i] in ['F', 'F S']:
            color_arr.append('gold')
        elif pos_array[i] in ['F M S', 'F M'] :
            color_arr.append('dodgerblue')
        elif pos_array[i] in ['M', 'M S']:
            color_arr.append('xkcd:salmon')
        else:
            color_arr.append('lawngreen')
    return np.array(color_arr)
```

```python
def ratio_perc(x, y):
    if x>0.3 or y>0.45:
        return True
    return False;
```
